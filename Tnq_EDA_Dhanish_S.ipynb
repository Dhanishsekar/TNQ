```python
df_features = df.drop(['Rank','CCA3','Country','Capital','Continent'],axis=1)
```

```python
df_features.boxplot()
plt.title('Distribution of all Numeric Variables', fontsize = 15)
plt.xticks(rotation = 'vertical', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="gnuplot_r",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,40))
ax.scatter(df['Growth Rate'],df['Country'])
ax.set_ylabel('Country')
ax.set_xlabel('Growth Rate')
plt.show()
```

```python
sns.countplot(df['Area (km²)'])
fig = plt.gcf()
fig.set_size_inches(6,6)
plt.title('Area (km²)')
```

```python
df.hist(xrot = 20,figsize = (15,15),color = "Red")
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='RdBu')
```

```python
df.groupby('Continent').Continent.count().plot(kind='pie')
plt.title('Continent')
plt.show()
```
